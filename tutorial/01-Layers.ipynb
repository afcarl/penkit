The most primitive data structure in penkit is the **layer**. Layers are just a pair of 1-dimensional numpy arrays of the same length. These two arrays represent the x and y coordinates of points in a plot, respectively. Points where both coordinates are `nan` represent discontinuities in the plot, i.e. places where the pen should be lifted between movements.

Here's an example of generating a layer with penkit:

```python
from penkit.textures import make_lines_texture
make_lines_texture(3, 2)
```

We can use the `preview` module to see what the layer looks like:

```python
from penkit.preview import show_layer
show_layer(make_lines_texture(3, 2))
```

As you can see, `make_lines_texture` generates a number of lines. **Textures** are a special type of layer where both dimensions fit into the range `(0, 1)`. These are useful when we get into surface projection. For now, though, we can just treat them as layers.

Let's look at some more textures:

```python
from penkit import textures
```

```python
show_layer(textures.make_grid_texture(5, 5))
```

```python
show_layer(textures.make_hex_texture(20, 20))
```

```python
show_layer(textures.make_spiral_texture(5))
```

In penkit, **plots** are just a list of layers that share a coordinate space. Instead of `show_layer` we can use `show_plot` to plot multiple layers. In pen plotting, multiple layers mean the plotter will return to a home location to give you a chance to change the writing instrument.

```python
from penkit.preview import show_plot
show_plot([
    textures.make_spiral_texture(6, 1000),
    textures.make_hex_texture(6)
])
```

To write a plot to an SVG file, you can use `penkit.write.write_plot`. By default, this will fit the plot to a US-letter sized page, adding a slight margin and preserving aspect ratio.

```python
from penkit.write import write_plot
```

```python
write_plot([textures.make_hex_texture(6)], 'spiral.svg')
```
